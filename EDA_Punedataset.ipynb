```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Puneclean12.csv")
```

```python
data
```

```python
data.columns
```

```python
data.info()
```

```python
# Line plot = Price by Balconies

sns.lineplot(data = data , x = "balconies" , y = "price", hue= "Neworold", marker="*")
plt.title("Price by Balconies")
plt.grid()
```

```python
# Barplot
# Price by Opensides

sns.barplot(data = data, x="opensides", y="price", hue="Neworold")
plt.title("Price by Opensides")
```

```python
# Bar plot
# Villa by price
sns.barplot(data=data, x="Villa", y="price", hue=data['Wifi'])
plt.title("Price By Villa with Wifi")
```

```python
# scatter plot
# Price by floor with bathroom
plt.figure(figsize=(8,10))
plt.scatter(data['Floor'], data['price'], c=data['bathroom'])
plt.colorbar()
plt.xlabel("Floor")
plt.ylabel("Price")
plt.title("Floor by Price with Bathroom")
plt.show()
```

```python
# Scatter Plot
# Price by Carpet Area with neworold

plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x="Carpetarea", y="price", hue="Neworold",palette="viridis")
plt.title("Price by Carpet Area")
plt.xlabel("Carpetarea")
plt.ylabel("Price")
plt.grid()
```

```python
# Subplot
# Price wise ares, Floors wise Bathrooms
plt.figure(figsize=(8,25))
fig, ax=plt.subplots(1,2)
ax[0].scatter(data["Area"], data["price"], )
ax[0].grid()
ax[0].set_xlabel("Area")
ax[0].set_ylabel("Price")
ax[0].set_title("Price wise ares")

ax[1].bar(data["bathroom"], data["Floor"])
ax[1].grid()
ax[1].set_xlabel("Bathroom")
ax[1].set_ylabel("Floor")
ax[1].set_title("Floors wise Bathrooms")
plt.show()
```

```python
# Histplot
# Carpetarea Distribution

plt.hist(data["Carpetarea"],bins=[1000,1500,2000,2500,3000,3500,4000,4500,5000])
plt.title("Carpetarea Distribution")
plt.xlabel("Carpetarea")
plt.ylabel("Count")
plt.grid()
```

```python
# Relplot = Lineplot
# Facing by price
sns.relplot(data=data, x="Facing", y="price", kind="line", hue="Neworold")
plt.xlabel("Facing")
plt.ylabel("price")
plt.xticks(rotation=60)
plt.title("Facing by price")
plt.grid()
```

```python
# Relplot = Scatterplot
# Facing by price
sns.relplot(data=data, x="ownership", y="price", kind="scatter")
plt.xlabel("Ownership")
plt.ylabel("price")
plt.xticks(rotation=10)
plt.title("Ownersip by price")
plt.grid()
```

```python
# Displot = kdeplot
# Facing by price
sns.displot(data=data, x="bathroom", kind="hist", fill=True)
plt.xlabel("Bathroom")
plt.title("Bathroom Distrubution")
plt.grid()
```

```python
# Kde plot
# Area Distribution
sns.kdeplot(data["Area"], fill=True)
plt.title("Area Distribution")
```

```python
# Displot = histplot
# Balconies Distribution
sns.histplot(data=data, x="balconies",fill=True)
plt.xlabel("Balconies")
plt.title("Balconies Distribution")
plt.grid()
```

```python
# Regression Plots
# regplot
sns.regplot(data=data, x="Floor", y="price")
```

```python
# residplot
sns.lmplot(data=data, x="Floor", y="price")
```
