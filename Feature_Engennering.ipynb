```python
# balconies nan value fill with 0
# mean median constant
# constant = 0
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
missing_constant= Pipeline(
    steps= [
        ("inpute", SimpleImputer(strategy= "constant", fill_value = 0))
    ]
)

missing_median= Pipeline(
    steps= [
        ("inpute", SimpleImputer(strategy= "median", add_indicator=True))
    ]
)

cat_ordinal= Pipeline(
    steps= [
        ("encode", OrdinalEncoder(categories=[["Old", "New"]]))
    ]
)

missing_mode_ordinal= Pipeline(
    steps= [
        ("inpute", SimpleImputer(strategy="constant", fill_value="No")),
        ("encode", OrdinalEncoder(categories=[["No", "Yes"],["No", "Yes"]]))
    ]
)

transformer = ColumnTransformer(
    transformers= [
        ("missing_constant_0", missing_constant, ["balconies","No of rooms"]),
        ("missing_median", missing_median, ["bathroom","Area", "Total rooms"]),
        ("cat_ordinal", cat_ordinal, ["Neworold"]),
        ("missing_mode_ordinal", missing_mode_ordinal, ["Car Parking", "Power Backup"])
    ],
    remainder = "passthrough"
)
```

```python
transformer.fit(xtrain)
```

```python
xtrain_t= transformer.transform(xtrain)
xtest_t= transformer.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
rf = RandomForestRegressor()
```

```python
ytrain, xtrain_t
```

```python
ytrain_t= np.log(ytrain)
ytest_t= np.log(ytest)
```

```python
rf.fit(xtrain_t, ytrain_t)
ypred=rf.predict(xtest_t)
```

```python
from sklearn.metrics import mean_absolute_error , r2_score
```

```python
r2_score(ytest_t, ypred)*100 , mean_absolute_error(ytest_t, ypred)*100
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
missing_value_constant=Pipeline(
    steps =[
            ("imput", SimpleImputer(strategy="constant", fill_value= 0))
       ]
)

missing_value_median = Pipeline(
    steps = [
        ("impute", SimpleImputer(strategy= "median",add_indicator=True))
    ]
)

categorical_ordinal = Pipeline(
    steps = [
        ("encode", OrdinalEncoder(categories=[["Old", "New"]]))
    ]
)

categorical_missing_ordinal = Pipeline(
    steps = [
        ("imput", SimpleImputer(strategy="constant", fill_value="No")),
        ("encode", OrdinalEncoder())
    ]
)


transformer= ColumnTransformer(
    transformers = [
        ("pipe1", missing_value_constant, ['balconies', 'No of rooms']),
        ("pipe2", missing_value_median, ['bathroom', 'Area', 'Total rooms']),
        ("pipe3", categorical_ordinal, ['Neworold']),
        ("pipe4", categorical_missing_ordinal, ['Car Parking', 'Power Backup', 'AC', 'TV','Refrigerator','Sofa','Washing Machine','Gas connection','BED','Wardrobe'])
    ]
)
```

```python
transformer.fit(xtrain)
```

```python
xtrain_t = transformer.transform(xtrain)
xtest_t = transformer.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import SVR
```

```python
rf= XGBRegressor()
```

```python
rf.fit(xtrain_t, ytrain_t)
ypred= rf.predict(xtest_t)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```
