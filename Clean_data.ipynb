```python
def clean_servent(x):
  x=str(x)
  if x == "nan":
    return np.nan
  elif "servant" in x:
    return "Yes"
  else:
    return "No"
data["Servant room"]=data["additionalrooms"].apply(clean_servent)

def clean_study(x):
  x=str(x)
  if x == "nan":
    return np.nan
  elif "study" in x:
    return "Yes"
  else:
    return "No"
data["Study room"]=data["additionalrooms"].apply(clean_study)
```

```python
def clean_age(x):
  x=str(x)
  if x== "nan":
    return np.nan
  elif "-" not in x:
    return x.split(" ")[0]
  else:
    x = x.split("-")
    a = x[0].strip()
    b = x[-1].strip().split(" ")[0]
    return (int(a) + int(b)) / 2
```

```python
data["Age"]=data["age"].apply(clean_age)
```

```python
def claen_bhk(x):
  x.strip().split(" ")
  if x[0].isdigit():
    a = x[1]
    num = x[0]
    if a == "BHK":
      if int(num) == 0:
        return 3
      else:
        return 3 + (int(num) - 1)
    else:
      return 2
  else:
    return np.nan
```

```python
data["Total rooms"]=data["bhk"].apply(claen_bhk)
```

```python
data[data["bhk"] == "1 BHK studio apartment "]
```

```python
def check_apperment(x):
  x=str(x)
  if "apartment" in x:
    return "Yes"
  else:
    return "No"
data["Aparment"]=data["bhk"].apply(check_apperment)
```

```python
def clean_bhk_residential(x):
  if 'residential plot' in x.lower():
    return 'Yes'
  else:
    return 'No'

def check_penthouse(x):
    x = str(x)
    if "penthouse" in x:
        return "Yes"
    else:
        return "No"

def check_Independent(x):
    x = str(x).lower()
    return "Yes" if "independent" in x else "No"

def check_villa(x):
    x = str(x)
    if "Villa" in x:
        return "Yes"
    else:
        return "No"

data["Villa"] = data["bhk"].apply(check_villa)
data["Independent"] = data["bhk"].apply(check_Independent)
data["Penthouse"] = data["bhk"].apply(check_penthouse)
data["Residential"] = data["bhk"].apply(clean_bhk_residential)
```
